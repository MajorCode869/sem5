```python
import requests
# change the method of your API key
from kaggle_secrets import UserSecretsClient
user_secrets = UserSecretsClient()
secret_value_0 = user_secrets.get_secret("api-key")


API_key=secret_value_0
base_url='https://api.themoviedb.org/3/'
```

```python
query=base_url+'/genre/movie/list?api_key='+API_key
response=requests.get(query)
```

```python
def get_data(genre_id,year,pageno,API_key):
    query=base_url+'/discover/movie?api_key='+API_key+'&with_genres='+str(genre_id)+'&primary_release_year.gte='+str(year)+'&sort_by=popularity.desc'+'&page='+str(pageno)
    response=requests.get(query)
    if response.status_code==200:
        data=response.json()
        return data['results']
    else:
        return 'error'
    
    
```

```python
get_data(35,2000,1,API_key)
```

```python
movies_list=[]
genre_id=35
year=2000
pageno=1
while(len(movies_list)<300):
    query_result=get_data(genre_id,year,pageno,API_key)
    if query_result=='error':
        continue
    else:
         for movie in query_result:
            movies_list.append(movie)
    pageno+=1    
```

```python
columns=['movie_id','movie_name','overview']
df=pd.DataFrame(columns=columns)
for movie in movies_list:
    df.loc[len(df)]=[movie['id'],movie['title'],movie['overview']]
```

```python
def get_similar(API_key):
    similar_movies=[]
    for id in df['movie_id']:
        query=base_url+"/movie/"+str(id)+"/similar?api_key="+API_key+'&page=1'
        response=requests.get(query)
        if(response.status_code==200):
            data=response.json()
            movies_list=data['results']
            sim=[]
            for i in range(0,5):
                sim.append(data['results'][i]['id'])
            similar_movies.append(sim);     
    
    
        else:
            continue
    return similar_movies        
    
```
